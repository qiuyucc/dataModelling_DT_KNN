```python
#Data Modelling -classification
#K-nearest neighbour
#Predict student's G3(final grade) by using students information
dfm = df.copy()
dfm = dfm.drop(['G3'],axis=1)
```

```python
dfm.head()
```

```python

#standardize the variables
#label encode grade
#0-Distinction, 2-Pass,  3-Fail
from sklearn import preprocessing
label = preprocessing.LabelEncoder()
dfm.Grade = label.fit_transform(dfm.Grade)
dfm.head()
```

```python
#train dataset & test dataset split #data prepartion 20% test dat
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
X=dfm.drop('Grade',axis=1)
y=dfm.Grade
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
#get dummy varibles
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
#check how many features in this dataset
len(list(X_train))
```

```python
#choosing a k -value   !elbow method
# create a for loop that trains various KNN models with different K values
# then keep track of the error rate for each of these models.
import numpy as np
error_rate = []
for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
plt.figure(figsize=(15,6))
plt.plot(range(1,30),error_rate,color='red',linestyle='dashed',marker='o', markerfacecolor='blue', markersize='10')
plt.title("Error rate by K-value while 0.2 test size")
plt.xlabel('no. of K')
plt.ylabel('Error Rate')
```

```python
#retrain with new k value, we can tell the best K value is 6 which has the least error rate
from sklearn.metrics import classification_report
clf = KNeighborsClassifier(n_neighbors =11)
fit =clf.fit(X_train, y_train)
y_pred = fit.predict(X_test)
cm =confusion_matrix(y_test, y_pred)
ts =fit.score(X_test,y_test)
print("Decisioin Tree Model Score" , ":" , fit.score(X_train, y_train) , "," , 
      "Cross Validation Score" ,":" , fit.score(X_test, y_test))
print ts
print cm
print(classification_report(y_test, y_pred))
```

```python
#decision tree test size =0.2
#find minimum samples leaf
from sklearn.tree import DecisionTreeClassifier 
msl=[]
for i in range(1,58):
    tree = DecisionTreeClassifier(min_samples_leaf=i)
    t= tree.fit(X_train, y_train)
    ts=t.score(X_test, y_test)
    msl.append(ts)
msl = pd.Series(msl)
msl.where(msl==msl.max()).dropna()
```

```python
# final model
import pydotplus
from sklearn.externals.six import StringIO 
from IPython.display import Image  
from sklearn.tree import export_graphviz
tree = DecisionTreeClassifier(min_samples_leaf=19)
t= tree.fit(X_train, y_train)
y_pred = t.predict(X_test)
cm =confusion_matrix(y_test, y_pred)
ts =t.score(X_test,y_test)
print ts
print cm
print(classification_report(y_test, y_pred))
print("Decisioin Tree Model Score" , ":" , t.score(X_train, y_train) , "," , 
      "Cross Validation Score" ,":" , t.score(X_test, y_test))
dot_data = StringIO()
export_graphviz(t,out_file=dot_data,filled=True, rounded =True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('DesicionTree20%.png')
Image(graph.create_png())
```
